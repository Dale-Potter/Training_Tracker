Training Tracker, to understand my progression to help train smarter.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('FitNotes Data/FitNotes_Export.csv')
```

```python
data.head()
```

```python
data = data.drop(['Distance Unit','Distance','Time'], axis = 1)
data['Volume'] = data['Weight (kgs)'] * data['Reps']
```

```python
data.head()
```

```python
Deadlift = data.loc[data['Exercise'] == 'Deadlift']
BackSquat = data.loc[data['Exercise'] == 'Barbell Squat']
BenchPress = data.loc[data['Exercise'] == 'Flat Barbell Bench Press']
```

```python
Deadlift.describe()
```

```python
BackSquat.describe()
```

```python
BenchPress.describe()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
Deadlift.groupby('Date').sum()['Volume'].plot(ax=ax)
BackSquat.groupby('Date').sum()['Volume'].plot(ax=ax)
BenchPress.groupby('Date').sum()['Volume'].plot(ax=ax)
ax.legend(['Deadlift Volume','Squat Volume','Bench Volume'])
```
