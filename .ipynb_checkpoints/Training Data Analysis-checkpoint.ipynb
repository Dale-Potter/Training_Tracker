Training Tracker, to understand my progression to help train smarter.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('FitNotes Data/FitNotes_Export.csv')
```

```python
data.head()
```

```python
data = data.drop(['Distance Unit','Distance','Time'], axis = 1)
data['Volume'] = data['Weight (kgs)'] * data['Reps']
```

```python
data.head()
```

```python
Conventional = data.loc[data['Exercise'] == 'Deadlift']
Snatch = data.loc[data['Exercise'] == 'Snatch Grip Deadlift']
Stiff = data.loc[data['Exercise'] == 'Stiff-Legged Deadlift']
Trap = data.loc[data['Exercise'] == 'Trap Bar Deadlift']
Sumo = data.loc[data['Exercise'] == 'Sumo Deadlift']
Deadlift = pd.concat([Conventional,Snatch,Stiff,Trap,Sumo])
Deadlift.sort_values('Date')
Deadlift.reset_index(inplace = True)
Deadlift = Deadlift.drop('index',axis = 1)
```

```python
BackSquat = data.loc[data['Exercise'] == 'Barbell Squat']
SafetySquat = data.loc[data['Exercise'] == 'Safety Squat Bar']
FrontSquat = data.loc[data['Exercise'] == 'Barbell Front Squat']
HighBarSquat = data.loc[data['Exercise'] == 'High Bar Squat']
Wraps = data.loc[data['Exercise'] == 'Squats With Wraps']
Squat = pd.concat([BackSquat,SafetySquat,FrontSquat,HighBarSquat,Wraps])
Squat.sort_values('Date')
Squat.reset_index(inplace = True)
Squat = Squat.drop('index', axis = 1)
```

```python
BenchPress = data.loc[data['Exercise'] == 'Flat Barbell Bench Press']
```

```python
fig, ax = plt.subplots(figsize=(15,7))
Deadlift.groupby('Date').sum()['Volume'].plot(ax=ax)
Squat.groupby('Date').sum()['Volume'].plot(ax=ax)
BenchPress.groupby('Date').sum()['Volume'].plot(ax=ax)
ax.legend(['Deadlift Volume','Squat Volume','Bench Volume'])
```
